```python
import numpy as np
import matplotlib.pyplot as plt
import wave
%matplotlib inline
# 该代码请参考https://zhuanlan.zhihu.com/p/19759362
```

```python
def showwav(path):
    wf=wave.open(path,"rb")
    params=wf.getparams()
    fr,fw=params[2],params[3]
    datawav=wf.readframes(fw)
    datause = np.fromstring(datawav,dtype = np.short)
    x=wf.getnchannels()
    datause.shape = -1,x
    datause = datause.T
    t = np.arange(0, fw) * (1.0/fr)
    plt.plot(t,datause[0])
    plt.show()
    xf=np.fft.fft(datause[0])
    xf_abs=np.fft.fftshift(abs(xf))
    print(len(xf_abs))
    plt.plot(t,xf_abs)
    plt.show()
```

```python
showwav("test1.wav")
```

```python
showwav("test.wav")
```
