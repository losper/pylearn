```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def plot_sin(A,f,phi):
    t=np.arange(0,2,0.01)
    y=A*np.sin(2*np.pi*t*f+phi)
    plt.plot(t,y)
    plt.show()
plot_sin(1,1,0)
```

```python
plot_sin(2,1,0)
```

```python
plot_sin(1,2,0)
```

```python
plot_sin(1,1,0)
```

```python
plot_sin(1,1,np.pi/2)
```
