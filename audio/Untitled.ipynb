```python
from matplotlib.mlab import specgram
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

t=np.arange(-1,1,0.001)
y=np.cos(2*np.pi*t)
z=np.cos(5*np.pi*t+3.14)
Fs=int(1.0/dt)

# df=specgram(y,Fs=Fs)
fig,axes=plt.subplots(nrows=1,ncols=2)

axes[0].specgram(x=y,Fs=Fs);
axes[1].specgram(x=z,Fs=Fs);

plt.show()
```
