```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
# 该代码请参考https://zhuanlan.zhihu.com/p/19759362
```

```python
# 画一个余弦波
t=np.arange(-1,1,0.01)
y=np.cos(2*np.pi*t)
plt.plot(t,y)
plt.show()
```

```python
# 叠加三个余弦波函数
y=np.cos(2*np.pi*t)+np.cos(3*2*np.pi*t)+np.cos(3*2*np.pi*0)
plt.plot(t,y)
plt.show()
```

```python
# 画直线波形
t=np.arange(-1,1,0.01)
y=np.sin(0*t)
# plt.ylim((-1, 1))
plt.plot(t,y)
plt.show()
```

```python
# 改变形状，只能通过波形函数相加来改变
y=(4*np.sin(3*2*np.pi*t))/(3*np.pi)+(4*np.sin(2*np.pi*t))/np.pi
plt.plot(t,y)
plt.show()
```

```python
# 改变形状，只能通过波形函数相加来改变
y=(4*np.sin(2*np.pi*t))/np.pi+(4*np.sin(3*2*np.pi*t))/(3*np.pi)+(4*np.sin(5*2*np.pi*t))/(5*np.pi)
plt.plot(t,y)
plt.show()
```

```python
xf=np.fft.fft(y)
xf_abs=np.fft.fftshift(abs(xf))
axis_xf=np.linspace(-1,1,200)
plt.plot(axis_xf,xf_abs)
plt.show()
```

```python
# 改变形状，只能通过波形函数相加来改变
x1 = np.random.rand(200)
y=x1+(4*np.sin(2*np.pi*t))/np.pi+(4*np.sin(3*2*np.pi*t))/(3*np.pi)+(4*np.sin(5*2*np.pi*t))/(5*np.pi)
plt.plot(t,y)
plt.show()
```

```python
# 注意看到下面频谱图的特点，注意振幅，频率，相位的不同引起的特点
xf=np.fft.fft(y)
xf_abs=np.fft.fftshift(abs(xf))
axis_xf=np.linspace(-1,1,200)
plt.plot(axis_xf,xf_abs)
plt.show()
```
